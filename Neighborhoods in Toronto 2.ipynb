Obtaining the data:

```python
data = pd.read_html("https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M")
df = data[0]
df.head()
```

Ignoring cells with a borough that is not assigned(i also renamed Postcode to PostalCode for continuity):

```python
df.rename(columns={"Postcode":"PostalCode"}, inplace=True)
df = df[df["Borough"]!="Not assigned"]
```

Merging neighborhoods in the same postal code area with a comma:

```python
df = df.groupby("PostalCode").agg(lambda x:", ".join(set(x)))
df.reset_index(inplace = True)
```

If a cell has a borough but is in a not assigned neighborhood, then the neighborhood will be the same as the borough

```python
df.loc[df["Neighbourhood"] == "Not assigned", "Neighbourhood"] = df.loc[df["Neighbourhood"] == "Not assigned", "Borough"]
df
```
