## Introduction/Business Problem

A finance company issues loans to people looking to buy a house. The compancy can offer loans for houses in urban, semi urban and rural areas. In order for a customer to obtain a loan, they must first apply for a home loan and then after that the company will validate the customer's eligibility for a loan.

The company wants an automatatic loan eligibility process based on the customer's details that they provide on their website. Each custome must fill in the following information into an online application form:
+ Gender
+ Marital status (i.e. if they are married or not)
+ Number of dependents (i.e how many other people that will be living with the customer)
+ Education	(i.e. are they a graduate or not)
+ Employment (i.e. are they self employed or not)
+ Applicant Income
+ Co-applicant Income
+ Loan Amount	
+ Loan Amount Term	
+ Credit History	
+ Property Area (i.e. in an urban, rural or semiurban area)

To automate this validity process, the company have given a problem to identify the customers information sections. The customers that are eligible for a loan can be easily identified so the finance company can specifically target these customers.

## Data
